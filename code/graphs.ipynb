```python
def draw_swarm_plot(dout2, name, categories, isColor=True):
    sns.set_style('whitegrid')
    fig, ax = plt.subplots(figsize=figure_dimensions)
    g=sns.swarmplot(x=[i[0] for i in dout2], y=[i[1] for i in dout2], palette=newPal,  size=6, 
                   order = categories
                   );
    
    if isColor:
        g.set(xlim=(0, 360))
        ax.set_xlabel('Hue')
        ax.set_ylabel('Term')
    else:
        g.set(xlim=(0, 1))
        ax.set_xlabel('Probability')
        ax.set_ylabel('Term')
    g.grid(True)
    g.set_title(name)
    #g.set(yticklabels=[y])
    sns.set(font_scale=2)

    plt.show()
```

```python
def draw_violin_plot(dout2, color):
    fig, ax = plt.subplots(figsize=figure_dimensions)
    for data in dout2:
        if data[1] == 'Maroon':
            if data[0] > 310:
                data[0] -= 360
    g = sns.violinplot(x=[i[0] for i in dout2], y=[i[1] for i in dout2], palette=newPal,
                   order = ["Reject","Maroon", "Brown","Olive","DarkGreen","DarkBlue", "DarkPurple"]
                   );
    g.set_title(color +  " Removed")
    g.set(xlim=(-50, 360))
    plt.show()
```

```python
def draw_box_plot(dout2, color):
    fig, ax = plt.subplots(figsize=figure_dimensions)
    for data in dout2:
        if data[1] == 'Maroon':
            if data[0] > 310:
                data[0] -= 360
    if (color == 'Maroon'):
        for data in dout2:
            if data[1] == 'Reject':
                if data[0] > 310:
                    data[0] -= 360
    g = sns.boxplot(x=[i[0] for i in dout2], y=[i[1] for i in dout2], palette=newPal,
                   order = ["Reject","Maroon", "Brown","Olive","DarkGreen","DarkBlue", "DarkPurple"]
                   );
    g.set(xlim=(-50, 360))
    g.set_title(color +  " Removed")
    plt.show()
```

```python
def reject_outliers(data, m=3):
    npdata = np.array(data)
    return npdata[abs(npdata - np.mean(npdata)) < m * np.std(npdata)]

def getDataPoints(color_data):
    hues = [[0],[0],[0],[0],[0],[0],[0],[0]]
    for hue in hues:
        hue.clear()
    for data in color_data:
        if data[1] == 'Reject':
            hues[0].append(data[0])
        elif data[1] == 'Red':
            hues[1].append(data[0])
        elif data[1] == 'Orange':
            hues[2].append(data[0])
        elif data[1] == 'Yellow':
            hues[3].append(data[0])
        elif data[1] == 'Green':
            hues[4].append(data[0])
        elif data[1] == 'Blue':
            hues[5].append(data[0])
        elif data[1] == 'Purple':
            hues[6].append(data[0])
        elif data[1] == 'Pink':
            hues[7].append(data[0])
    for x in range(len(hues)):
        hues[x] = reject_outliers(hues[x])
    return hues
```

```python
colors = ['none','red','orange','yellow','green','blue','purple','pink']

for x in range(len(all_data)):
    print('\n' + colors[x], 'removed')
    color_data = getDataPoints(all_data[x])
    for y in range(len(color_data)):
        print(colors[y])
        newcolor = np.array(color_data[y])
        if len(newcolor) >0:
            print(np.mean(newcolor), np.std(newcolor), np.percentile(newcolor, 50))
        else:
            print("N/A")
```
